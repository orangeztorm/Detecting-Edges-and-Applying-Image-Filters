# BLURRING

Blurring is reffered to as averaging the pixel values in a neigborhood. This is also called a low pass filter. A low pass filter is a filter which allows low frequencies and block high frequencies.
A frequency in an image mean rate of change of pixel in an image.
A simle way to pass a low pass filter is by uniformly averaging the values in the neighborhood of a pixel. We can choose the kernel on how much we want to smoothen the image.

```python
# Sample code for blurring an image
# Importing importing the important libraries
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Read in an image
image = cv.imread('pexel-7.jpg')

# Converting from BGR to RGB
image = cv.cvtColor(image, cv.COLOR_RGB2BGR)

#
plt.figure(figsize=(10,10))

# visualize your image
plt.imshow(image)
```

```python
rows, cols = image.shape[:2]

# Generating identity kernel
kernel_identity = np.array([[0, 0, 0], [0, 1, 0], [0, 0, 0]])

# Generating a 3x3 low pass filter
kernel_3x3 = np.ones((3,3), np.float32) / 9.0

# Generating a 5x5 low pass filter
kernel_5x5 = np.ones((5,5), np.float32) /25.0
```

```python
# Passing the in the image , ddepth and kernel
# ddepth refers to desired depth of the destination image
# When ddepth=-1, the output image will have the same depth as the source.

kernel_identity_output = cv.filter2D(image, -1, kernel_identity)

#
plt.figure(figsize=(10,10))

# There will be no visible change in the image as shown below because it's an identity matrix
plt.imshow(kernel_identity_output)
```

```python
# For the 3x3
output_3x3_kernel_image = cv.filter2D(image, -1, kernel_3x3)

#
plt.figure(figsize=(10,10))

# It not obvious but it's a little blurry
plt.imshow(output_3x3_kernel_image)
```

```python
# for 5x5 kernel
output_5x5 = cv.filter2D(image, -1, kernel_5x5)

# 
plt.figure(figsize=(10,10))

#
plt.imshow(output_5x5)
```

An optional way of doing this will be using the cv2.blur

```python
output = cv.blur(image, (9,9))

plt.figure(figsize=(10, 10))

plt.imshow(output)
```

# Edge Detection

The process of edge detection involves detecting sharp edges in an image producing a binary image as the output. What we do is that we draw white or black lines  to indicate such egdes in an image. Edge detection is an high pass filtering operation, so we can say it the reverese of blurring where in blurring we block high frquencies and allow low frequencies, but in edgde detection we do the reverese. 

```python
# Using sobel filter

img = cv.imread('soduku.jpg')

#
img = cv.cvtColor(img, cv.COLOR_RGB2BGR)

# Converting to gray scale
image_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# This will will show only horizontal edges
sobel_horizontal = cv.Sobel(image_gray, cv.CV_64F, 1, 0, ksize=5 )

# This will show vertical lines
sobel_vertical = cv.Sobel(image_gray, cv.CV_64F, 0, 1, ksize=5 )
```

```python
# Output for sobel vertical
plt.figure(figsize=(10,10))
plt.imshow(sobel_vertical, cmap='gray')
```

```python
# Output for sobel horizontal
plt.figure(figsize=(10,10))
plt.imshow(sobel_horizontal, cmap='gray')
```

```python
# To combine both or show both combied 
laplacain = cv.Laplacian(image_gray, cv.CV_64F)

plt.figure(figsize=(10,10))
plt.imshow(laplacain, cmap='gray')
```

```python
# using canny edge
canny = cv.Canny(image_gray, 50, 240)

#
plt.figure(figsize=(10, 10))
plt.imshow(canny)
```
